```python
df[['PRICEEACH', 'FLAG']]
```

```python
def foo(x):
    if x < 10:
        return 'BAD'
    elif x > 10 and x < 50:
        return 'GOOD'
    else:
        return 'EXCEL'

df['WORTH'] = df[['QUANTITYORDERED']].applymap(foo)
```

```python
df[['QUANTITYORDERED', 'WORTH']]
```

```python
def ftrust(x, y):
    if x <= y:
        return 'YES'
    else: 
        return 'NO'

df['TRUST'] = list(map(ftrust, df['PRICEEACH'], df['MSRP']))
```

```python
df[['PRICEEACH', 'MSRP', 'TRUST']]
```

```python
dict_map = {1: 'Q1', 2: 'Q2', 3: 'Q3', 4: 'Q4'}
df['QTR_ID'] = df['QTR_ID'].map(dict_map)
```
